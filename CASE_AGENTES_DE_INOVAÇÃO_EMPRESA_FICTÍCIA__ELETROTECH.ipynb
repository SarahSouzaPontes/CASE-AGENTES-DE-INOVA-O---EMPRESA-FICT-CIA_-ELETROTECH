```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fictícios de vendas de três produtos ao longo do tempo
dados = {
    'Mes': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    'Produto_A': [100, 120, 90, 110, 130, 150, 140, 160, 170, 180, 200, 190],
    'Produto_B': [80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190],
    'Produto_C': [120, 130, 140, 150, 160, 170, 180, 190, 200, 210, 220, 230]
}

# Converter para DataFrame
df = pd.DataFrame(dados)

# Calcular a média móvel para suavizar os dados e identificar padrões sazonais
janela = 3  # Janela de média móvel
df['Produto_A_MM'] = df['Produto_A'].rolling(window=janela).mean()
df['Produto_B_MM'] = df['Produto_B'].rolling(window=janela).mean()
df['Produto_C_MM'] = df['Produto_C'].rolling(window=janela).mean()

# Plotar gráfico de linhas das vendas dos três produtos e suas médias móveis
plt.figure(figsize=(10, 6))
plt.plot(df['Mes'], df['Produto_A'], marker='o', label='Produto A')
plt.plot(df['Mes'], df['Produto_A_MM'], linestyle='--', label='Produto A (Média Móvel)')
plt.plot(df['Mes'], df['Produto_B'], marker='o', label='Produto B')
plt.plot(df['Mes'], df['Produto_B_MM'], linestyle='--', label='Produto B (Média Móvel)')
plt.plot(df['Mes'], df['Produto_C'], marker='o', label='Produto C')
plt.plot(df['Mes'], df['Produto_C_MM'], linestyle='--', label='Produto C (Média Móvel)')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.title('Vendas de Produtos ao Longo do Tempo - CASE AGENTES DE INOVAÇÃO - EMPRESA FICTÍCIA: ELETROTECH')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords

# Download dos stopwords da língua portuguesa (caso não tenham sido baixados)
nltk.download('stopwords')

# Obter lista de stopwords da língua portuguesa
stop_words = set(stopwords.words('portuguese'))

# Depoimentos dos clientes
depoimentos = [
    "O smartphone é incrível! A câmera tira fotos de alta qualidade e o desempenho é ótimo.",
    "Estou muito satisfeito com o tablet. A bateria dura bastante e o design é muito elegante.",
    "A qualidade do notebook é decepcionante. A tela apresenta problemas de visualização e o teclado é desconfortável.",
    "O smartwatch é excelente! As funções são muito úteis e o design é moderno.",
    "O tablet é ruim. A tela fica travando e o aplicativo de navegação é lento.",
    "O notebook atendeu às minhas expectativas. É leve e fácil de transportar, mas a duração da bateria poderia ser melhor."
]

# Extrair palavras dos depoimentos e remover stopwords
palavras = [word.lower() for depoimento in depoimentos for word in depoimento.split() if word.lower() not in stop_words]

# Contar a frequência das palavras
contagem_palavras = Counter(palavras)

# Palavras mais frequentes
top_palavras = contagem_palavras.most_common(30)

# Criar a nuvem de palavras
nuvem_palavras = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(top_palavras))

# Plotar a nuvem de palavras
plt.figure(figsize=(10, 6))
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.title('Nuvem de Palavras dos Depoimentos dos Clientes (sem stopwords)')
plt.axis('off')
plt.show()
```
